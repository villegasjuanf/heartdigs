```python
import wget
import pandas as pd

#repository URL 
PATH='https://raw.githubusercontent.com/villegasjuanf/heartdigs/master'

#function to download files from repo
def get_file(path,filename):
    if filename not in os.listdir():
        try:
            wget.download(os.path.join(path,filename))
        except:
            return None
            print('archivo %s inexistente'%filename)
        else:
            return filename
            print('file %s downloaded'%filename)

#file index table downloading
get_file(PATH,'set_a.csv') 
```

```python
set_a=pd.read_csv('set_a.csv')
set_a['filename']=set_a.fname.apply(lambda x: x[6::])

train_set=set_a.filename[set_a.label.notnull()]
print(train_set)

#audio files download
wav_files=[]
for wav_file in train_set:
    f=get_file(PATH,wav_file)
    if f is not None:
        wav_files.append(f)
```

```python
from scipy.io import wavfile
import numpy as np

data=dict()
f=dict()

print('archivo | dimension | frec muestreo')

#wave files to Numpy
for file in wav_files:
    f_,data_=wavfile.read(file)
    f[file]=f_ #dict of  sample frequencies
    data[file]=np.array(data_) #data dict
    print(file,data_.shape[0],f_)
```

```python
from scipy.signal import decimate

DOWNSAMPLE=60 # resampling to fs/DOWNSAMPLE
N_DATA=4000*120//120 # crop data to first N_DATA
CREATE_M=True #create a new DFT matrix M (use TRUE once if you change N_DATA or DOWNSAMPLE)

#resampling and padding to fit each data serie to N_DATA lenght
data_dec=dict()
fs=dict()
for i in wav_files:
    fs[i]=f[i]/DOWNSAMPLE
    data_dec[i]= decimate(data[i],q=DOWNSAMPLE,ftype='fir',zero_phase=True)[0:N_DATA] 
    if len(data_dec[i])<N_DATA:
        data_dec[i]=np.append(data_dec[i],[0]*(N_DATA-len(data_dec[i])))
    print("%s  n: %d  fs: %0.1f"%(i,len(data_dec[i]),fs[i]))
```

```python
#funcion de calculo de la DFT
def dft_systemml(signal):
    prog = dml(dml_dft).input('signal', signal).output('DFT')
    
    return (

    #execute the script inside the SystemML engine running on top of Apache Spark
    ml.execute(prog) 
     
         #read result from SystemML execution back as SystemML Matrix
        .get('DFT') 
     
         #convert SystemML Matrix to ApacheSpark DataFrame 
        .toNumPy() 
    )
```

```python
#DFT calculation for the all the audio data
data_dft=dict()
for i,file in enumerate(wav_files):
    data_=np.array(data_dec[file]).reshape([N_DATA,1])
    print("%d/%d  %s: (%d,%d)\n"%(i+1,len(wav_files),file,data_.shape[0],data_.shape[1]))
    data_dft[file]=dft_systemml(data_)
```

```python
#serialization functions to save the data set

import json
def DataSerialize(data_dft,data_dec,fs):
    toJsonF=dict()
    toJsonT=dict()
    for key in data_dft.keys():
        toJsonF[key]=data_dft[key].reshape([len(data_dft[key])]).tolist()
        toJsonT[key]=data_dec[key].reshape([len(data_dec[key])]).tolist()
    toJsonF=json.dumps(toJsonF)
    toJsonT=json.dumps(toJsonT)
    toJsonS=json.dumps(fs)
    toJson=json.dumps({'data_f':toJsonF,'data_t':toJsonT,'fs':toJsonS})
    return toJson

def DataDeserialize(jsonObj):
    toDict=json.loads(jsonObj)
    data_f=json.loads(toDict['data_f'])
    data_t=json.loads(toDict['data_t'])
    fs=json.loads(toDict['fs'])
    for key in data_f.keys():
        data_f[key]=np.array(data_f[key]).reshape([len(data_f[key]),1])
        data_t[key]=np.array(data_t[key])
    
    return data_f,data_t,fs
```

```python
#data set storage and reading
WRITE_DATASET=False
READ_DATASET=False

if WRITE_DATASET:
    with open('dataset.json','w') as f:
        f.write(DataSerialize(data_dft,data_dec,fs))
    print([i for i in os.listdir() if 'dataset.json' in i])
if READ_DATASET:
    with open('dataset.json','r') as f:
        dataset=f.read()
    data_dft,data_dec,fs=DataDeserialize(dataset)
```
